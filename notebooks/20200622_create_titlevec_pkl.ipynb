```python
import pandas as pd
import yaml
```

```python
CONFIG_FILE = '../configs/config.yaml'
with open(CONFIG_FILE) as file:
    yml = yaml.safe_load(file)

MODEL_DIR_NAME = yml['SETTING']['MODEL_DIR_NAME']
SUB_DIR_NAME = yml['SETTING']['SUB_DIR_NAME']
RAW_DATA_DIR_NAME = yml['SETTING']['RAW_DATA_DIR_NAME']
FEATURE_DIR_NAME = yml['SETTING']['FEATURE_DIR_NAME']
```

```python
train = pd.read_pickle(FEATURE_DIR_NAME + 'train.pkl')
test = pd.read_pickle(FEATURE_DIR_NAME + 'test.pkl')
```

```python
train.head()
```

```python
title_vec_col = [col for col in train.columns if col.startswith('title_vec')]
```

```python
train_title = train[title_vec_col]
test_title = test[title_vec_col]
```

```python
title_vec = pd.concat([train_title, test_title], axis=0)
```

```python
title_vec
```

```python
title_vec = title_vec.reset_index(drop=True)
```

```python
title_vec
```

```python
title_vec.to_pickle(FEATURE_DIR_NAME + 'title_vec.pkl')
```

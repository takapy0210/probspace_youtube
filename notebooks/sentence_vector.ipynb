```python
import pandas as pd
import numpy as np
from bert_sentence_vectorizer import BertSequenceVectorizer
```

```python
train = pd.read_csv('../data/raw/train_data.csv', index_col=0)
```

```python
train.head()
```

```python
train_title = train[['video_id', 'title']]
```

```python
train_title = train_title.head(20).reset_index(drop=True)
```

```python
train_title
```

```python
BSV = BertSequenceVectorizer()
```

```python
# train_title['title_feature'] = train_title['title'].progress_apply(lambda x: BSV.vectorize(x))
title_feature = train_title['title'].progress_apply(lambda x: BSV.vectorize(x))
```

```python
title_feature = pd.DataFrame(title_feature.values.tolist())
title_feature = title_feature.add_prefix('title_vec')
```

```python
title_feature
```

```python
train_title = pd.concat([train_title, title_feature], axis=1)
```

```python
import pandas as pd
import yaml
import numpy as np

import umap
import bhtsne
```

```python
CONFIG_FILE = '../configs/config.yaml'
with open(CONFIG_FILE) as file:
    yml = yaml.safe_load(file)

MODEL_DIR_NAME = yml['SETTING']['MODEL_DIR_NAME']
SUB_DIR_NAME = yml['SETTING']['SUB_DIR_NAME']
RAW_DATA_DIR_NAME = yml['SETTING']['RAW_DATA_DIR_NAME']
FEATURE_DIR_NAME = yml['SETTING']['FEATURE_DIR_NAME']
```

```python
title_vec = pd.read_pickle(FEATURE_DIR_NAME + 'title_vec.pkl')
```

```python
title_vec
```

```python
%%time
# UMAP
um = umap.UMAP(n_components=10, random_state=42)
_umap = um.fit_transform(title_vec)
umap_df = pd.DataFrame(_umap, columns=['title_umap1', 'title_umap2', 'title_umap3', 'title_umap4', 'title_umap5', 'title_umap6', 'title_umap7', 'title_umap8', 'title_umap9', 'title_umap10'])
```

```python
umap_df
```

```python
umap_df.to_pickle(FEATURE_DIR_NAME + 'title_umap_10d.pkl')
```

```python
%%time
# t-SNE
_bhtsne = bhtsne.tsne(title_vec.astype(np.float64), dimensions=2, rand_seed=42)
bhtsne_df = pd.DataFrame(_bhtsne, columns=['title_tsne1', 'title_tsne2'])
```

```python
bhtsne_df
```

```python
_df = pd.concat([umap_df, bhtsne_df], axis=1)
```

```python
_df
```

```python
_df.to_pickle(FEATURE_DIR_NAME + 'wave_umap_tsne.pkl')
```
